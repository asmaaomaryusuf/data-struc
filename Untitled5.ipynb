```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
from sklearn.neighbors import KNeighborsClassifier # KNN 
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV # train - test split and fine tuning for KNN
from sklearn.preprocessing import MinMaxScaler # Norm
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, roc_curve, roc_auc_score
```

ùêÉùêöùê≠ùêö ùê¨ùê©ùê•ùê¢ùê≠

```python
Norm = MinMaxScaler(feature_range = (0, 1))
x = Data.drop(["Outcome"], axis = 1)
x = Norm.fit_transform(x)
y = Data["Outcome"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 45)
```

ùêäùêçùêç

```python
ModelKNN = KNeighborsClassifier()
```

```python
K_Values = np.array([11, 12, 13, 14, 15, 16, 17, 18])
metric = ["minkowski", "chebyshev"]
p = np.array([5, 6, 7, 8, 9, 10, 11])
param_grid = {"n_neighbors": K_Values, "metric": metric, "p": p}
GridKNN = GridSearchCV(estimator = ModelKNN, param_grid = param_grid, cv = 5)
GridKNN.fit(x_train, y_train)
```

```python
print(f"B.score = {GridKNN.best_score_}")
print(f"k = {GridKNN.best_estimator_.n_neighbors}")
print(f"metric = {GridKNN.best_estimator_.metric}")
print(f"p = {GridKNN.best_estimator_.p}")
```

```python
ModelKNN = KNeighborsClassifier(n_neighbors = 15, metric = "minkowski", p = 9)
ModelKNN.fit(x_train, y_train)
```

ùêèùê´ùêûùêùùê¢ùêúùê≠ùê¢ùê®ùêß

```python
y_predKNN = ModelKNN.predict(x_test)
```

ùêÇùê®ùêßùêüùêÆùê¨ùê¢ùê®ùêß ùê¶ùêöùê≠ùê´ùê¢ùê± 

```python
MatrixKNN = confusion_matrix(y_test, y_predKNN)
print(MatrixKNN)
```

```python
plt.figure(figsize = (10,7))
sbn.heatmap(MatrixKNN, annot = True, vmin = 0, vmax = 150, cmap = "coolwarm")
plt.show()
```

```python
print(f"Accuracy KNN: {accuracy_score(y_test, y_predKNN)}")
print(f"Recall KNN: {recall_score(y_test, y_predKNN)}")
print(f"Precision KNN: {precision_score(y_test, y_predKNN)}")
```

ùêçùêöùê¢ùêØùêû ùêÅùêöùê≤ùêûùê¨

```python
ModelNB = GaussianNB()
ModelNB.fit(x_train, y_train)
```

ùêèùê´ùêûùêùùê¢ùêúùê≠ùê¢ùê®ùêß

```python
y_predNB = ModelNB.predict(x_test)
```

ùêÇùê®ùêßùêüùêÆùê¨ùê¢ùê®ùêß ùê¶ùêöùê≠ùê´ùê¢ùê±

```python
MatrixNB = confusion_matrix(y_test, y_predNB)
print(MatrixNB)
```

```python
plt.figure(figsize = (10,7))
sbn.heatmap(MatrixNB, annot = True, vmin = 0, vmax = 150, cmap = "coolwarm")
plt.show()
```

```python
print(f"Accuracy Naive Bayes: {accuracy_score(y_test, y_predNB)}")
print(f"Recall Naive Bayes: {recall_score(y_test, y_predNB)}")
print(f"Precision Naive Bayes: {precision_score(y_test, y_predNB)}")
```
